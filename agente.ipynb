```python
import urllib.request as ur
import re
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

request = ur.Request("http://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000")
result = ur.urlopen(request)
content = result.read().decode()



exp_tabla = '.*?<table.*?id="ctl00_Contenido_tblAcciones".*?".*?>(.*?)</table>.*?'
tabla = re.findall(exp_tabla,content,re.DOTALL)
#Hecho
exp_compañias ='.*?<td.*?align="left">.*?<a.*?>(.*?)</a>*?'
#Hecho
exp_ultimos ='.*?<td class="DifFl.*?" align="left">.*?<td>(.*?)</td>.*?'
#Hecho
exp_diferencias = '.*?<td>.*?</td>.*?<td class="DifCl.*?">(.*?)</td>.*?<td>.*?</td>.*?'
#Hecho
exp_minimos = '.*?<td>([0-9]+,[0-9]+)</td><td>[0-9]+\.[0-9]+.*?</td>.*?'
#Hecho
exp_maximos = '.*?<td class="DifCl.*?">.*?<td>(.*?)</td>.*?'
#Hecho
exp_efectivos = '<td>[0-9]+\..*?[0-9]+.>*?</td>.*?<td>([0-9]+\.[0-9]+,[0-9]+)</td>.*?'
#Hecho
exp_volumen = '<td>([0-9]+?\.?[0-9]+\.[0-9]+)</td>.*?<td>[0-9]+\.[0-9]+,[0-9]+</td>.*?'
#Hecho
exp_fechas = '.*?<td>[0-9]+\.[0-9]+,[0-9]+</td>.*?<td.*?>(.*?)</td>.*?'
#Hecho
exp_horas = '.*?<td class="Ult" align="center">(.*?)</td>.*?</tr>'

#Listas
compañias = []
ultimos = []
diferencias = []
maximos = []
minimos = []
efectivos = []
volumenes = []
fechas = []
horas = []

#Funciones
def getCompañias(compañias):
    compañias.append(re.findall(exp_compañias,content,re.DOTALL))
    compañias[0].pop(0)
    
def getUltimos(ultimos):
    ultimos.append(re.findall(exp_ultimos,''.join(tabla),re.DOTALL))    
    
def getDiferencias(diferencias):
    diferencias.append(re.findall(exp_diferencias,''.join(tabla),re.DOTALL))

def getMaximos(maximos):
    maximos.append(re.findall(exp_maximos,''.join(tabla),re.DOTALL))

def getMinimos(minimos):
    minimos.append(re.findall(exp_minimos,''.join(tabla),re.DOTALL))
    
def getEfectivo(efectivos):
    efectivos.append(re.findall(exp_efectivos,''.join(tabla),re.DOTALL))

def getVolumen(volumenes):
    volumenes.append(re.findall(exp_volumen,''.join(tabla),re.DOTALL))
    
def getFecha(fechas):
    fechas.append(re.findall(exp_fechas,''.join(tabla),re.DOTALL))
    
def getHoras(horas):
    horas.append(re.findall(exp_horas,''.join(tabla),re.DOTALL))

    #Obtención de los datos de la tabla de acciones de la Bolsa de Madrid
#def obtenerDatosAcciones():
getCompañias(compañias)
getUltimos(ultimos)
getDiferencias(diferencias)
getMaximos(maximos)
getMinimos(minimos)
getEfectivo(efectivos)
getVolumen(volumenes)
getFecha(fechas)
getHoras(horas)

#def ListadeListatoLista():
compañia = compañias.pop(0)
ultimo = ultimos.pop(0)
diferencia = diferencias.pop(0)
maximo = maximos.pop(0)
minimo = minimos.pop(0)
efectivo = efectivos.pop(0)
volumen = volumenes.pop(0)
fecha = fechas.pop(0)
hora = horas.pop(0)

#def pandaToExcel():
 #Creamos los pandas para visualizar los datos
acciones = {'Compañía' : compañia ,'Último' : ultimo ,'Diferencia': diferencia ,'Máximo':maximo ,'Mínimo':minimo,'Efectivo' : efectivo,'Volumen':volumen,'Fecha': fecha,'Hora':hora} 
df = pd.DataFrame.from_dict(acciones)
writer = ExcelWriter('Acciones1.xlsx')
df.to_excel(writer,'Sheet1',index=False)
writer.save() 
df

    

 
   
```
